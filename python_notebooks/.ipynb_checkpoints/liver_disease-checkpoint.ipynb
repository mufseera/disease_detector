```python
import numpy as np
import pandas as pd
import random
import seaborn as sns
from sklearn.model_selection import train_test_split
import pickle
import warnings
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/Users/mufseeramusthafa/Documents/machine_learning/multi_disease_detection/dataset/liver.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
sns.countplot("Dataset",data=df)
plt.show()
```

```python
df["Dataset"].replace([2,1],[1,0],inplace=True)
```

```python
sns.countplot("Dataset",data=df)
plt.show()
```

```python
sns.countplot("Gender",data=df,hue="Dataset")
plt.show()
```

```python
df=pd.get_dummies(df,columns=["Gender"],drop_first=True)
```

```python
df.head()
```

```python
corr=df.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.pairplot(df,hue="Dataset")
plt.show()
```

```python
x=df.drop(columns=["Dataset"])
y=df["Dataset"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,test_size=.15)
```

```python
model=RandomForestClassifier(n_estimators=20)
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
import pickle 
pickle.dump(model,open("liver.pkl","wb"))
```
